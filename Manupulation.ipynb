```python
def pred(source):
    from ultralytics import YOLOv10
    model = YOLOv10.from_pretrained('jameslahm/yolov10n')
    results = model.predict(source=source)
    for result in results:
    # Now we can iterate through detections in each result
        for det in result.boxes:
        # Get normalized xywh (YOLO format) and class confidence
            X = det.xywhn
            conf = det.conf
            cls = det.cls
        # Print in the desired format
            print(f"x_center: {X}")
            if(torch.any(torch.isnan(X))):
              return(np.zeros([1,4]))
            else:
              return(X)
```

```python
def merge_csv(a):

    import pandas as pd
    import glob
    import os
    path2='/content/drive/MyDrive/Colab Notebooks/Untitled folder/result/beam.csv'
    df1=pd.read_csv(path2,dtype=np.float64)
    for i in range(1,len(a)):
        df2=pd.read_csv(str(a[i]),dtype=np.float64)
        combined_df = pd.concat([df1, df2.transpose()], ignore_index=True)
        combined_df.to_csv('combined_file.csv', index=False)
        df1=combined_df
    return combined_df
```
